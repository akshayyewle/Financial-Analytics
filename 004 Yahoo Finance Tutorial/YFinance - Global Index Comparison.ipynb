```python
import yfinance as yf
import matplotlib.pyplot as plt
```

# **Import Data**

```python
tickerSymbol = '^FTSE'
tickerData = yf.Ticker(tickerSymbol)
tickerDf = tickerData.history(period='20y')

plt.plot(tickerDf['Close'])
plt.title('LSE Index')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

# **Compare**

```python
Period = '25y'
```

```python
FTSE100 = yf.Ticker('^FTSE').history(period=Period)

NASDAQ100 = yf.Ticker('^NDX').history(period=Period)
sp500Data = yf.Ticker('^GSPC').history(period=Period)
dowData = yf.Ticker('^DJI').history(period=Period)

Sensex_IN = yf.Ticker('^BSESN').history(period=Period)
NIFTY50 = yf.Ticker('^NSEI').history(period=Period)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(FTSE100['Close'], label='FTSE 100')
plt.plot(NASDAQ100['Close'], label='NASDAQ 100')
plt.plot(sp500Data['Close'], label='S&P 500')
plt.plot(dowData['Close'], label='Dow Jones Industrial Average')

plt.plot(Sensex_IN['Close'], label='Sensex_IN')
plt.plot(NIFTY50['Close'], label='NIFTY50')

plt.title('Index Comparison')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
