# **Setup**

```python
import pandas as pd 
from datetime import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as pdr
plt.style.use('fivethirtyeight')
```

# **Data Import**

```python
Start_Date = datetime(2010,1,1)
End_Date = datetime.today()
```

# **CPI Data**

```python
# https://fred.stlouisfed.org/
# https://fred.stlouisfed.org/series/CORESTICKM159SFRBATL

df = pdr.DataReader(name='CORESTICKM159SFRBATL',data_source='fred',start=Start_Date,end=End_Date)
df.head(10)
```

```python
df.plot(kind='line',figsize=(8,4),
        title='Sticky Price Consumer Price Index less Food and Energy')
plt.show()
```

# **GDP Data**

```python
df = pdr.DataReader(name='GDP',data_source='fred',start=Start_Date,end=End_Date)
display(df.head(5))

df.plot(kind='line',figsize=(8,4),title='Gross Domestic Product (BUSD)')
plt.show()
```
